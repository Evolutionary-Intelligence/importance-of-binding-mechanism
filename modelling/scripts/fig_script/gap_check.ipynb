```python
import numpy as np
import os
import pandas as pd
import plotly.graph_objects as go

import modelling
```

```python
SA_model = modelling.SensitivityAnalysis()
```

```python
fig = go.Figure()

# Read data for drugs
saved_data_dir = '../../../simulation_results/'
filename = 'parameter_space_gaps_res5.csv'
df = pd.read_csv(saved_data_dir + filename,
                 header=[0], index_col=[0],
                 skipinitialspace=True)

Vhalf_list = df['Vhalf'].values
Kmax_list = df['Kmax'].values
Ku_list = df['Ku'].values

# Read data for space
saved_data_dir = '../../../simulation_data/sensitivity_analysis/'
file = 'parameter_space_res5.csv'

df = pd.read_csv(saved_data_dir + file,
                 header=[0], index_col=[0],
                 skipinitialspace=True)

Vhalf_range = df['Vhalf'].values
Kmax_range = df['Kmax'].values
Ku_range = df['Ku'].values


fig.add_trace(
    go.Scatter3d(
        x=Vhalf_list,
        y=Kmax_list,
        z=Ku_list,
        mode='markers',
        name='gaps',
        marker=dict(
            color='red',
            size=5,
        )
    )
)

fig.add_trace(
    go.Scatter3d(
        x=Vhalf_range,
        y=Kmax_range,
        z=Ku_range,
        mode='markers',
        name='space',
        marker=dict(
            color='blue',
            size=5,
        )
    )
)

fig.update_layout(scene = dict(
                    xaxis_title='Vhalf',
                    yaxis_title='Kmax',
                    zaxis_title='Ku',
                    yaxis = dict(dtick=1,
                                 type='log'),
                    zaxis = dict(dtick=1,
                                 type='log')))

fig.show()
```
